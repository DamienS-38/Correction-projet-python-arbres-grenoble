## Manipulation des colonnes

```python
# Création d'un dataframe réduit avec uniquement 3 colonnes
df2 = df[['GENRE_BOTA', 'ESPECE', 'ANNEEDEPLANTATION']].copy()
df2.sample(10)
```

```python
# Création d'une colonne supplémentaire
df2["GENRE_ESPECE"] = df2.GENRE_BOTA + ' ' + df2.ESPECE
df2
```

```python
# Création d'un nouveau dataframe en enlevant les NaN (Not a Number)
df2 = df2.dropna(axis=0).copy()
df2
```

```python
df2.nunique()
```

```python
# On récupère le niombre d'éléments uniques 
# En faisant attention à ne pas avoir de faux doublons 
(df2.GENRE_ESPECE.str.lower()
                 .str.strip()
                 .nunique()
)
```
