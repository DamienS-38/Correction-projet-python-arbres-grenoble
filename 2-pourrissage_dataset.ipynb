```python
import numpy as np
import pandas as pd
```

```python
# Chargement des donn√©es dans un dataframe pandas
df = pd.read_csv('data/arbres.csv')
```

```python
def change_name_prefix(df):
    nbre_changement = 50
    for i in range(nbre_changement):
        index = np.random.randint(len(df))
        if type(df.loc[index, 'GENRE_BOTA']) is str:
            df.loc[index, 'GENRE_BOTA'] = ''.join(["Ceci est un : ", df.loc[index, 'GENRE_BOTA']])

def add_unnecessary_space(df):
    nbre_changement = 50
    for i in range(nbre_changement):
        index = np.random.randint(len(df))
        num_space = np.random.randint(5)
        if type(df.loc[index, 'ESPECE']) is str:
            df.loc[index, 'ESPECE'] = ''.join([" "*num_space, df.loc[index, 'ESPECE']])   
            
def change_case(df):
    nbre_changement = 50
    for i in range(nbre_changement):
        index = np.random.randint(len(df))
        if type(df.loc[index, 'GENRE_BOTA']) is str:
            df.loc[index, 'GENRE_BOTA'] = df.loc[index, 'GENRE_BOTA'].upper()
```

```python
df = pd.read_csv('data/arbres.csv')
change_name_prefix(df)
add_unnecessary_space(df)
change_case(df)

df.to_csv('data/weird_trees.csv')
```
